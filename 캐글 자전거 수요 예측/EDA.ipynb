```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df['year'] = pd.to_datetime(df['datetime']).dt.year
```

```python
df['month'] = pd.to_datetime(df['datetime']).dt.month
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
```

```python
df
```

```python
fig, ax = plt.subplots(3, 3, figsize=(10, 10), gridspec_kw={'hspace': 0.5, 'wspace': 0.5})

sns.set_palette('deep')

sns.barplot(x='temp', y='count',data=df, ax=ax[0,0], palette='deep')
sns.barplot(x='season', y='count',data=df, ax=ax[1,0], palette='deep')
sns.barplot(x='holiday', y='count',data=df, ax=ax[2,0], palette='deep')
sns.barplot(x='workingday', y='count',data=df, ax=ax[0,1], palette='deep')
sns.barplot(x='weather', y='count',data=df, ax=ax[1,1], palette='deep')
sns.barplot(x='atemp', y='count',data=df, ax=ax[2,1], palette='deep')
sns.barplot(x='humidity', y='count',data=df, ax=ax[0,2], palette='deep')
sns.barplot(x='windspeed', y='count',data=df, ax=ax[1,2], palette='deep')
sns.barplot(x='casual', y='count',data=df, ax=ax[2,2], palette='deep')
```

```python
matplotlib.rc('font', size=15)
sns.displot(df['count'], palette='deep')
```

```python
import numpy as np
```

```python
matplotlib.rc('font', size=15)
sns.displot(np.log1p(df['count']), palette='deep')
```

```python
df.select_dtypes(include=int).columns
```

```python
matplotlib.rc('font', size=14)
matplotlib.rc('axes', titlesize=15)
figure, axes = plt.subplots(nrows=3, ncols=2, figsize=(10,9), gridspec_kw={'hspace':0.5, 'wspace':0.5})
plt.tight_layout()

sns.barplot(x='year', y='count', data=df, ax=axes[0,0])
sns.barplot(x='month', y='count', data=df, ax=axes[1,0])
sns.barplot(x='holiday', y='count', data=df, ax=axes[2,0])
sns.barplot(x='weather', y='count', data=df, ax=axes[0,1])
sns.barplot(x='workingday', y='count', data=df, ax=axes[1,1])

axes[0,0].set(title='Rental amounts by year')
axes[1,0].set(title='Rental amounts by month')
axes[2,0].set(title='Rental amounts by holiday')
axes[0,1].set(title='Rental amounts by weather')
axes[1,1].set(title='Rental amounts by workingday')

axes[1,0].tick_params(axis='x', labelrotation=90)
axes[2,0].tick_params(axis='x', labelrotation=90)
```

```python
figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10), gridspec_kw={'hspace':0.5, 'wspace':0.5})
plt.tight_layout()

sns.boxplot(x='season', y='count', data=df, ax=axes[0, 0])
sns.boxplot(x='weather', y='count', data=df, ax=axes[0, 1])
sns.boxplot(x='holiday', y='count', data=df, ax=axes[1, 0])
sns.boxplot(x='workingday', y='count', data=df, ax=axes[1, 1])

axes[0, 0].set(title='Box Plot On Count Across Season')
axes[0, 1].set(title='Box Plot On Count Across Weather')
axes[1, 0].set(title='Box Plot On Count Across Holiday')
axes[1, 1].set(title='Box Plot On Count Across Workingday')
```

```python
df['hour'] = pd.to_datetime(df['datetime']).dt.hour
```

```python
df['weekday'] = pd.to_datetime(df['datetime']).dt.weekday
```

```python
matplotlib.rc('font', size=11)
figure, axes = plt.subplots(nrows=5, ncols=1)
figure.set_size_inches(12,18)

sns.pointplot(x='hour', y='count', data=df, hue='workingday', ax=axes[0])
sns.pointplot(x='hour', y='count', data=df, hue='holiday', ax=axes[1])
sns.pointplot(x='hour', y='count', data=df, hue='weekday', ax=axes[2])
sns.pointplot(x='hour', y='count', data=df, hue='season', ax=axes[3])
sns.pointplot(x='hour', y='count', data=df, hue='weather', ax=axes[4])
```

```python
matplotlib.rc('font', size=15)
figure, axes = plt.subplots(2,2)
plt.tight_layout()
figure.set_size_inches(7,6)

sns.regplot(x='temp', y='count', data=df, ax=axes[0,0], scatter_kws={'alpha':0.2}, line_kws={'color':'blue'})
sns.regplot(x='atemp', y='count', data=df, ax=axes[0,1], scatter_kws={'alpha':0.2}, line_kws={'color':'blue'})
sns.regplot(x='windspeed', y='count', data=df, ax=axes[1,0], scatter_kws={'alpha':0.2}, line_kws={'color':'blue'})
sns.regplot(x='humidity', y='count', data=df, ax=axes[1,1], scatter_kws={'alpha':0.2}, line_kws={'color':'blue'})
```
